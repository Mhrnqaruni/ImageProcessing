```python
# Import sympy library
import sympy as sp
x = sp.Symbol('x')

#real function
f = x

period = 100

#find the coefficient A
def calculateAk(k):
    fAk = f * sp.cos(k*x)
    #integrate between -pi and pi
    return sp.integrate(fAk, (x, -sp.pi, sp.pi))/sp.pi


#find the coefficient B
def calculateBk(k):
    fBk = f * sp.sin(k*x)
    #integrate between -pi and pi
    return sp.integrate(fBk, (x, -sp.pi, sp.pi))/sp.pi

fourier_series = calculateAk(0)/2
# sum of the functions (using cos and sin)
for k in range(period+1):
    fourier_series += calculateAk(k+1)*sp.cos((k+1)*x)
    fourier_series += calculateBk(k+1)*sp.sin((k+1)*x)

# Plot the graph of f(x)
sp.plot(fourier_series, (x, -sp.pi, sp.pi), title='Fourier Series of f(x)', xlabel='x', ylabel='f(x)')
sp.plot(f, (x, -sp.pi, sp.pi), title='Real f(x)', xlabel='x', ylabel='f(x)')
```

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Define the variable and the function
x = sp.symbols('x')
f = x**3

# Define the limits
L = sp.pi  # You can adjust this value according to your function's period
k = sp.symbols('k', integer=True)

# Calculate the Fourier coefficients
Ck = (1/(2*L))*sp.integrate(f*sp.exp(-sp.I*k*sp.pi*x/L), (x, -L, L))

# Define the number of terms in the Fourier series
N = 10  # You can adjust this value

# Calculate the Fourier series
fourier_series = sum(Ck.subs(k, n)*sp.exp(sp.I*n*sp.pi*x/L) for n in range(-N, N+1))

# Lambdify the Fourier series for plotting
fourier_func = sp.lambdify(x, fourier_series, 'numpy')

# Generate x values
x_vals = np.linspace(float(-L), float(L), 400)

# Generate y values (Fourier series)
y_vals = fourier_func(x_vals)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='Fourier series of $x^2$')
plt.legend()
plt.grid(True)
plt.show()
```
